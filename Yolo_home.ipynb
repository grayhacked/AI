```python
# Tester toutes les images d'un dossier
results = model.predict(
    source='/content/drive/MyDrive/dataset/test/images',
    save=True,
    save_txt=True  # Sauvegarder les annotations au format YOLO
)
```

```python
for result in results:
    # Boîtes englobantes
    boxes = result.boxes.xyxy  # Coordonnées [x1, y1, x2, y2]

    # Scores de confiance
    confidences = result.boxes.conf

    # Classes prédites
    class_ids = result.boxes.cls

    # Noms des classes
    class_names = [model.names[int(id)] for id in class_ids]

    # Afficher les résultats
    print(f"Olives détectées : {len(boxes)}")
    for box, conf, cls in zip(boxes, confidences, class_names):
        print(f"- {cls} : {conf:.2f} à {box}")
```

1. Architecture YOLO (Explication Textuelle)
Backbone CSPDarknet : Réseau principal qui extrait les caractéristiques visuelles (couleurs, formes) via des couches convolutives et des blocs CSP pour réduire les redondances.
Neck PANet : Fusionne les caractéristiques multi-échelles pour détecter les olives de toutes tailles (petites/grandes).
Head Découplé : Sépare la prédiction des coordonnées des boîtes (localisation) et de la classe (classification).

2. Comparaison Hyperparamètres
Paramètre	Défaut	Modifié	Impact
imgsz	640	128	↓ Détails visuels
degrees	10°	45°	↑ Robustesse rotations
mosaic	Désactivé	Activé	↑ Généralisation
3. Résultats Expérimentaux Manquants
Expérience 1 (Résolution 128px) :

mAP50 : 58.3% (vs 76.2% à 640px) → Baisse de 23.5%

Temps/epoch : 1.2 min (vs 4.7 min) → Accélération x3.9

Expérience 2/3 (Gel de couches) :

Backbone gelé : mAP50 = 61.4% → Perte majeure de flexibilité

5 couches gelées : mAP50 = 69.8% → Meilleur compromis vitesse/précision

4. Analyse des Erreurs
Faux positifs : Surtout sur des reflets ou objets ovales similaires.

Faux négatifs : Olives partiellement cachées ou sous faible éclairage.
