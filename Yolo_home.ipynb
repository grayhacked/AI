```python
for result in results:
    # BoÃ®tes englobantes
    boxes = result.boxes.xyxy  # CoordonnÃ©es [x1, y1, x2, y2]

    # Scores de confiance
    confidences = result.boxes.conf

    # Classes prÃ©dites
    class_ids = result.boxes.cls

    # Noms des classes
    class_names = [model.names[int(id)] for id in class_ids]

    # Afficher les rÃ©sultats
    print(f"Olives dÃ©tectÃ©es : {len(boxes)}")
    for box, conf, cls in zip(boxes, confidences, class_names):
        print(f"- {cls} : {conf:.2f} Ã  {box}")
```
